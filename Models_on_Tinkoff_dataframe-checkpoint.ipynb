# Logistic Regression

```python
def logistic_regression(X, y):
    
    """
    Данная функция выполняет обучение логистической регрессии и измеряет необходимые метрики качеста.
    """
    # Инициализируем логистическую регрессию и проведем разбиение на train и test, после чего обучим ее
    lr = LogisticRegression()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
    lr.fit(X_train, y_train)
    
    # Сформируем предсказания
    predictions = lr.predict(X_test)
    # Измерим accuracy на train и test
    lr_acc_train = lr.score(X_train, y_train) 
    lr_acc_test = lr.score(X_test, y_test) 
    # Сформируем вероятность прогнозов
    y_scores = lr.predict_proba(X_test)
    y_scores = y_scores[:,1]
    
    # Выведем accuracy, precision, recall
    print("Logistic Regression accuracy on train:{:.3f}".format(lr_acc_train))
    print("Logistic Regression accuracy on test:{:.3f}".format(lr_acc_test))
    
    print("Precision on Logistic Regression:{:.3f}".format(precision_score(y_test, predictions)))
    print("Recall on  Logistic Regression:{:.3f}".format(recall_score(y_test, predictions)))
    
    # Оценим accuracy на 5-ти блочной кросс-валидации 
    cv_test_scores = cross_val_score(lr, X_test, y_test, cv=5, scoring = 'accuracy')
    print("Результаты accuracy на кросс-валидации:")
    print()
    print("Scores:", cv_test_scores)
    print("Standard Deviation:", cv_test_scores.std())
    print("Mean:", cv_test_scores.mean())
    print()
    # Оценим roc_auc на 5-ти блочной кросс-валидации
    cv_scores = cross_val_score(lr, X_test, y_test, cv=5, scoring = 'roc_auc')
    print("Результаты roc_auc на кросс-валидации:")
    print()
    print("Scores:", cv_scores)
    print("Standard Deviation:", cv_scores.std())
    print("Mean:", cv_scores.mean())
    print()
    print("AUC на тестовой выборке:{:.2f}".format(roc_auc_score(y_test, lr.predict_proba(X_test)[:, 1])))
    
    # Изобразим график roc_auc
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_scores)
    def plot_roc_curve(false_positive_rate, true_positive_rate, label=None):
        plt.plot(false_positive_rate, true_positive_rate, linewidth=2, label=label)
        plt.plot([0, 1], [0, 1], 'r', linewidth=4)
        plt.axis([0, 1, 0, 1])
        plt.xlabel('False Positive Rate (FPR)', fontsize=16)
        plt.ylabel('True Positive Rate (TPR)', fontsize=16)

    plt.figure(figsize=(14, 7))
    plot_roc_curve(false_positive_rate, true_positive_rate)
    plt.show()
    
    # Изобразим график PR - curve
    precision, recall, threshold = precision_recall_curve(y_test, y_scores)
    def plot_precision_and_recall(precision, recall, threshold):
        plt.plot(threshold, precision[:-1], "r-", label="precision", linewidth=5)
        plt.plot(threshold, recall[:-1], "b", label="recall", linewidth=5)
        plt.xlabel("threshold", fontsize=19)
        plt.legend(loc="upper right", fontsize=19)
        plt.ylim([0, 1])

    plt.figure(figsize=(14, 7))
    plot_precision_and_recall(precision, recall, threshold)
    plt.show()
```

### StandartScaler + RandomUnderSampler + LogisticRegression

```python
# В качестве первого алгоритма классификации я буду использовать логистическую регрессию, которая требует масштабирования данных
# Применим к датафрейму StandartScaler(), после чего решим проблему дисбаланса классов путем применения RandomUnderSampler.
stsc = StandardScaler()
# Отделим целевую переменную от всех остальных признаков
tink_features = tink_df.drop('open_account_flg', axis = 1)
result_value = tink_df['open_account_flg']
# Отмасштабируем данные
stsc_df = pd.DataFrame(stsc.fit_transform(tink_features), columns = tink_features.columns)
# Решим проблему дисбаланса
X_resampled, y_resampled = rus.fit_sample(stsc_df, result_value)
# Применим функцию по обучению логистической регрессии и посмотрим на результат
logistic_regression(X_resampled, y_resampled)
```

```python
def decision_tree_classifier(X, y):
    tree = DecisionTreeClassifier(max_depth = 5)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
    tree.fit(X_train, y_train)
    
    # Сформируем предсказания
    tree_predictions = tree.predict(X_test)
    # Измерим accuracy на train и test
    tree_acc_train = tree.score(X_train, y_train) 
    tree_acc_test = tree.score(X_test, y_test) 
    # Сформируем вероятность прогнозов
    y_scores = tree.predict_proba(X_test)
    y_scores = y_scores[:,1]
    
    # Выведем accuracy, precision, recall
    print("Decision tree accuracy on train:{:.3f}".format(tree_acc_train))
    print("Decision tree accuracy on test:{:.3f}".format(tree_acc_test))
    
    print("Precision on Decision tree:{:.3f}".format(precision_score(y_test, tree_predictions)))
    print("Recall on  Decision tree:{:.3f}".format(recall_score(y_test, tree_predictions)))
    
    # Оценим accuracy на 5-ти блочной кросс-валидации 
    cv_test_scores = cross_val_score(tree, X_test, y_test, cv=5, scoring = 'accuracy')
    print("Результаты accuracy на кросс-валидации:")
    print()
    print("Scores:", cv_test_scores)
    print("Standard Deviation:", cv_test_scores.std())
    print("Mean:", cv_test_scores.mean())
    tree_auc = cv_test_scores.mean()
    print()
    # Оценим roc_auc на 5-ти блочной кросс-валидации
    cv_scores = cross_val_score(tree, X_test, y_test, cv=5, scoring = 'roc_auc')
    print("Результаты roc_auc на кросс-валидации:")
    print()
    print("Scores:", cv_scores)
    print("Standard Deviation:", cv_scores.std())
    print("Mean:", cv_scores.mean())
    print()
    print("AUC на тестовой выборке:{:.2f}".format(roc_auc_score(y_test, tree.predict_proba(X_test)[:, 1])))
    
    # Изобразим график roc_auc
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_scores)
    def plot_roc_curve(false_positive_rate, true_positive_rate, label=None):
        plt.plot(false_positive_rate, true_positive_rate, linewidth=2, label=label)
        plt.plot([0, 1], [0, 1], 'r', linewidth=4)
        plt.axis([0, 1, 0, 1])
        plt.xlabel('False Positive Rate (FPR)', fontsize=16)
        plt.ylabel('True Positive Rate (TPR)', fontsize=16)

    plt.figure(figsize=(14, 7))
    plot_roc_curve(false_positive_rate, true_positive_rate)
    plt.show()
    
    # Изобразим график PR - curve
    precision, recall, threshold = precision_recall_curve(y_test, y_scores)
    def plot_precision_and_recall(precision, recall, threshold):
        plt.plot(threshold, precision[:-1], "r-", label="precision", linewidth=5)
        plt.plot(threshold, recall[:-1], "b", label="recall", linewidth=5)
        plt.xlabel("threshold", fontsize=19)
        plt.legend(loc="upper right", fontsize=19)
        plt.ylim([0, 1])

    plt.figure(figsize=(14, 7))
    plot_precision_and_recall(precision, recall, threshold)
    plt.show()
```
